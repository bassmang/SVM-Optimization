```python
import numpy as np
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = datasets.load_iris()
X = iris['data'];
y = iris['target'];
cols=['Sepal Length', 'Sepal Width', 'Pedal Length', 'Pedal Width']
X_pd = pd.DataFrame(data=X,columns=cols)
y_pd = pd.Series(data=y)
X_pd_all = X_pd.copy()
X_pd_all['target'] = y_pd
target = np.where(X_pd_all['target'], 'Versicolor', 'Setosa')
target[np.where(y == 2)] = 'Virginica'
X_pd_all['target description'] = target;
sns.pairplot(X_pd_all.drop(columns=['target']), hue='target description', height=2);
```
